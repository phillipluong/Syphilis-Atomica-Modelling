## Next, we aim to include a 'burn-in' period so that the time can reasonably cancel out before we do any calibration

```python
F = at.ProjectFramework('syphilis_neg_framework_v01.xlsx') # Let's create a new databook based on this framework
# D = at.ProjectData.from_spreadsheet('syphilis_neg_databook_v01_2.xlsx', framework=F)
# P = at.Project(framework=F, databook= D)
```

```python
D = at.ProjectData.new(framework=F, tvec=list(range(2009,2020)),pops=1,transfers=0)
D.save('syphilis_neg_databook_v02.xlsx')
```

```python
D = at.ProjectData.from_spreadsheet('syphilis_neg_databook_v02.xlsx',framework=F)
P = at.Project(framework=F, databook=D,sim_dt=1/365.25)
```

```python
d = at.PlotData(P.results[0],project=P2)
at.plot_series(d,plot_type='stacked',data=P.data);
```

```python
at.plot_series(d,data=P.data);
```

```python
with at.Quiet():
    p3 = P.calibrate(max_time=3000, parset='default', adjustables=['b_rate','doth_rate'],measurables=['sus','inf'])
```

```python
p3.pars['b_rate'].y_factor
```

```python
p3.pars['doth_rate'].y_factor
```

```python
r3= P.run_sim(parset=p3, result_name = 'Calibrated births')
d = at.PlotData([P.results[0],r3],outputs=['sus','inf'],project=P)
at.plot_series(d, axis='results', data=P.data);
```

#### Let's run this multiple times to see if repeated calibration works for this

```python
cal_par = P.parsets['default'].copy()
with at.Quiet():
    for _ in range(10):
        cal_par = P.calibrate(max_time=3000, parset=cal_par, adjustables=['doth_rate'], measurables = ['sus','inf'])
        cal_par = P.calibrate(max_time=3000, parset=cal_par, adjustables=['infx'], measurables = ['sus','inf'])
```

```python
cal_par.pars['infx'].y_factor
```

```python
cal_par.pars['doth_rate'].y_factor
```

```python
r4= P.run_sim(parset=cal_par, result_name = 'Calibrated births')
d = at.PlotData([P.results[0],r4],outputs=['sus','inf'],project=P)
at.plot_series(d, axis='results', data=P.data);
```
